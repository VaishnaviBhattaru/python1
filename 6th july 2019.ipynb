```
#print the largrst digit
def prinlarge(n):
  i=0
  while n!=0:
    r=n%10
    if i < r:
      i=r
    n=n//10
  return i
prinlarge(2168)
```

```
def ispalindrom(n):
  rev=0
  buffer=n
  while n!=0:
    x=n%10
    rev=rev*10+x
    n=n//10
  if buffer == rev:
    return"yes"
  else:
    return"no"
print(ispalindrom(121))
print(ispalindrom(123))
  
```

```
#for loop print n nos
def printnnos(n):
  for x in range(1,n+1):
    print(x,end=" ")
  return
printnnos(100)
  
```

```
def printnnosbw(n,y):
  for x in range(n,y+1):
    print(x,end=" ")
  return
printnnosbw(10,20)
```

```
#print alternative
def printaltnnosbw(n,y):
  for x in range(n,y+1,2):
    print(x,end=" ")
  return
printaltnnosbw(10,30)
```

```
#print rev
def printrev(n,y):
  for x in range(y,n-1,-1):
    print(x,end=" ")
  return
printrev(10,30)
```
