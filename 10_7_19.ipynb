```
contacts={}
def addcontacts(name,phone):
  if name not in contacts:
    contacts[name]=phone
    print("contact is added")
  else:
    print("contact already exists")
  return
addcontacts('anil','21684321894')
addcontacts('medha','23416548913')
addcontacts('jaggu','46588432156')
addcontacts('medha','32698744853')
```

```
contacts
```

```
def searchcontact(name):
  if name in contacts:
    print(name,":",contacts[name])
  else:
    print("%s does not exist"%name)
  return
searchcontact('jaggu')
searchcontact('charan')
```

```
def importcontacts(newcontacts):
  contacts.update(newcontacts)
  print(len(newcontacts.key()),"contacts added sucessfully")
  return
newcontacts={'charan':65648946894,'gowtham':26519481894}
importcontacts(newcontacts)
```

```
#3. The program must accept two string values S1 and S2. The program must print the pattern as shown in the 
#Example Input/Output sections.
def combi(s1,s2):
  l1 = len(s1)
  l2 = len(s2)
  if l2>l1:
    l = l2
    for x in range(l1,l+1):
      s1.append("*")
  else:
    l = l1
    for y in range(l2,l+1):
      s2.append("*")
  for i in range(0,l):
    
    print(s1[i],s2[i])
    print(" ")
  return

s1 = input("Enter string 1")
l1=[]
for i in range(0,len(s1)):
  l1.append(s1[i])
s2 = input("Enter string 2")
l2=[]
for i in range(0,len(s2)):
  l2.append(s2[i])

combi(l1,l2)
```
